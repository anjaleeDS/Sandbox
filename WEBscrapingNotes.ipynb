# *Beautiful* Soup intro

```
import bs4  #beautiful soup
import lxml.etree as xml
import requests  # go out and fetch itnative python
import pandas as pd
import numpy as np
import re
```

```
# url = "https://www.tradesy.com/collection/designer-drop/?brand=hermes&condition=new-with-tags|like-new&sort=featured"
url = "https://en.wikipedia.org/wiki/Artificial_intelligence"
requests.get(url)
```

```
requests.get(url, {}).text
```

```
soup = bs4.BeautifulSoup(requests.get(url, {}).text, 'lxml')
```

```
soup.title
```

```
soup.title.text
```

```
# soup.find()  # this is useful but find_all is even better
# soup.find_all('li')  #finds all <li> elements
# soup.find_all('li')[0]  # reference the first <li> element
```

```
regular_expression = re.compile('^tocsection-')
content_list = soup.find_all('li', attrs={'class': regular_expression}) # only list object that has tocsection as attribute
print(content_list)
```

```
content = []

for li in content_list:
  content.append(li.get_text().split('\n'))

content
```

```
# get to the "See Also" section and grab the links below it
# look for the table and get discerning attributes around it

see_also_section = soup.find('div', attrs={'class': 'div-col columns column-width'})
see_also_section
```

```
see_also_list_elems = see_also_section.find_all('li') # get the elements that have the tag <li>
see_also_list_elems
```

```
### this doesn't work!

# see_also_list_html = see_also_list_elems.find_all('a href=')
# see_also_list_html
```

```
see_also = []

for li in see_also_list_elems:
  a_tag = li.find('a', href=True)
  href = a_tag['href']
  see_also.append(href)

see_also
```

# Tradesy with beautiful soup

```
tradesy = "https://www.tradesy.com/search/?brand=hermes&condition=like-new&q=hermes%20birkin&num_per_page=192&sort=recently-listed"
```

```
requests.get(tradesy)

soupt = bs4.BeautifulSoup(requests.get(tradesy, {}).text, 'lxml')
soupt
```

```
soupt.title.text
```

```
# regex = re.compile("item-tile-full-price")
price_list = (soupt.find_all('span', attrs={'itemprop':"priceCurrency"}))
name_list = (soupt.find_all('div', attrs={'itemprop':"name"}))
link_list = (soupt.find_all('a', attrs={'class':"item-tile__title-link"}))
```

```
print("\n ... link  ", link_list[191], "\n... name ", name_list[191], "\n..  price .. ", price_list[191])
```

```
lot = zip(link_list, name_list, price_list)
export_file = pd.DataFrame(lot, columns=['links','name','price'])
```
